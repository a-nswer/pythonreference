# Jupyter Notebook

Jupyter Notebook is like a notebook that also runs code

```python
import os
os.listdir()
```

Jupyter notebook has a command mode (esc key to access) and an edit mode (highlight a cell and enter to access). 

Here are some important shortcuts:
- ctrl+enter will run a cell
- shift+enter will run a cell and select the next one
- alt+enter will run a cell and insert a new code cell underneath

When we're in command mode:
- we can change the type of a cell to markdown with the m key
- we can change the type of a cell to code with the y key

At any point, we can access the **command palette** with **ctrl+shift+f**

Tab completion is very comprehensive, in that it will basically explain everything to you. Press Tab to see a basic list of possible arguments, or you can press shift+tab more than once to see things in more detail each time, try it below:

```python
import numpy as np
x = np.arange(12)
print()
```

Jupyter notebook also has magic functions, they start with a %

```python
%time sum([x for x in range(100000)])
```

```python
%time sum(x for x in range(100000))
```

As you can see above, the `%time` magic function has showed us how much faster a generator comprehension summation can be versus a list comprehension summation. Below, we use `$matplotlib inline` to tell the notebook to display our plots here
