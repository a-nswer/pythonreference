## Pandas
In this notebook we look at pandas. Most of this is taken from the Pandas tutorial found here:
https://pandas.pydata.org/pandas-docs/stable/tutorials.html

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

We can use Pandas to load csv files, like the one in this directory

```python
df = pd.read_csv('comptagevelo2012.csv', encoding='latin1', index_col='Date', parse_dates=['Date'], dayfirst=True)
df[:3]
```

When you read a csv, you end up with a DataFrame object. A DataFrame object has rows and columns. You can access dataframe object columns the same way you would access a dictionary's elements

```python
df['Rachel / Papineau'][:5]
```

You can use matplotlib to plot these dataframes very easily

```python
df['Rachel / Papineau'].plot()
```
